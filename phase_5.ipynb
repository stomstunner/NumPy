## here we learn about array loading and saving in memory

```python
import numpy as np
import matplotlib.pyplot as plt
```

### sometime we want ki ham different array/ matrix ko memeory me save karwaye

```python
arr1 = np.array([
    [1,2,3],
    [4,5,6]
])
arr2 = np.random.rand(3,3)
arr3 = np.zeros((4,4))

# now we want ki ham isse save kar le memory me 
np.save('arr1.npy',arr1)
np.save('arr2.npy',arr2)
np.save('arr3.npy',arr3)
# yaha pe hame apne file ka naam jo ki .npy formate me hona chaiye likhna hoga aur kiss array ko ham usse file me covert karna chate hai usse likhna hoga 

# print("The saved array is: ",)

# first of all lets load the array in my file
loaded_array1  = np.load("arr1.npy")
print("The array 1 is : \n",loaded_array1)
```

```python
# lets do operation on numply logo 
try:
    logo  = np.load('numpy-logo.npy')
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy Logo")
    plt.grid(False)
except FileNotFoundError:
    print("The file is not found")
    
    
```

```python
# we can make the dark logo also
try:
    
    dark_logo = 1-logo
    plt.figure(figsize=(10,5))
    plt.subplot(121)
    plt.imshow(dark_logo)
    plt.title("Numpy Logo dark")
    plt.grid(False)
except FileNotFoundError:
    print("The file is not found")
    
  
```
